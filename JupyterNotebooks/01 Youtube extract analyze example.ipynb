# ROSÉ & Bruno Mars - APT.

# Download MP3 from youtube

```python
url = "https://www.youtube.com/watch?v=ekr2nIex040"
input_file = "rose1.m4a"
output_file = "rose1.wav"
yt = YouTube(url, on_progress_callback=on_progress)
print(f"Title: {yt.title}")
print(f"Author: {yt.author}")
ys = yt.streams.get_audio_only()
ys.download(filename=input_file)

# Convert M4A to WAV
os.system(f"ffmpeg -i {input_file} {output_file}")

#!ffmpeg -i rose1.wav # Print wav file metadata
```

## Analysis

```python
#  Essentia analysis
extractor = MusicExtractor()
features, _ = extractor(output_file)

# Print all available feature keys:
#print("Available keys in features:")
#for key in features.descriptorNames():
    #print(key)

print(f"Danceability: {features['rhythm.danceability']}")
print(f"Key Scale: {features['tonal.key_temperley.scale']}")
print(f"Average Loudness: {features['lowlevel.average_loudness']} dB")
```
